```
%pylab inline    
x = linspace(-10,10,300)         # 定義 x 從-10到10 產生300個點

plot(x,sin(x))                   # 畫 sin(x) 和 cos(x)
plot(x,cos(x))

plot(x[x>0],sin(x)[x>0],"b--")   # x>0 範圍的 sin(x) 藍色虛線
plot(x[x<0],cos(x)[x<0],"r:")   # x<0 範圍的 cos(x) 紅色點線
```
