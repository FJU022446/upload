```
#[] plot means of 4 traits in subplots
# 載入packages
%matplotlib inline
import numpy as np 
from matplotlib import pyplot as plt 
from sklearn.datasets import load_iris
#其他資料可參考 : https://scikit-learn.org/stable/datasets/index.html#toy-datasets

# 載入iris data
iris = load_iris()
X = iris.data
# 指定性狀名稱
trait_list = ['sepal_length','sepal_width','petal_length','petal_width']
class_names=iris.target_names

iris_3=iris.data.reshape((3,50,4))    #改變維度
iris_means = np.mean(iris_3,axis=1)   #計算50個值的平均數
iris_vars = np.var(iris_3,axis=1)     #計算50個值的變異數

print("The means are \n",iris_means,"\n")
print("The variances are \n",iris_vars,"\n")

print("The bar plot of means")

#畫圖
plt.figure(figsize=(10,10))          #設定畫布大小

for i in range((len(trait_list))):    #每個性狀畫一張
    plt.subplot(2,2,(i+1))
    plt.ylabel(trait_list[i])
    plt.bar(class_names,iris_means[:,i],color=["r","y","g"])
    
plt.savefig('C:\python\iris_bar.png')  #存檔(先產生 C:\python目錄)
```
